```python
import numpy as np
```

```python
# addition
arr1 = np.array([10,2,62,21])
arr2 = np.array([11,34,20,1])

arr3 = arr1 + arr2

print(arr3)
```

```python
arr3 = np.add(arr1,arr2) # we can add any number of array
print(arr3)
```

```python
# Multiplication

arr1 = np.array([10,2,62,21])
arr2 = np.array([11,34,20,1])

arr3 = arr1*arr2

print(arr3)
```

```python
arr3 = np.multiply(arr1,arr2)
print(arr3)
```

```python
# subtract


arr1 = np.array([10,2,62,21])
arr2 = np.array([11,34,20,1])

arr3 = arr1-arr2

print(arr3)
```

```python
arr3 = np.subtract(arr1,arr2)
print(arr3)
```

```python
# divide


arr1 = np.array([10,2,62,21])
arr2 = np.array([11,34,20,1])

arr3 = arr1/arr2

print(arr3)
```

```python
# Mod

arr3 = np.mod(arr1,arr2) # will give the reminder
print(arr3)
```

```python
# remainder
arr3 = np.remainder(arr1,arr2) # will give the reminder
print(arr3)
```

```python
# power

arr1 = np.array([2,3,4,5])
arr2 = np.array([3,2,4,2])

arr3 = np.power(arr1,arr2) # will give the reminder
print(arr3)
```

```python
# statistical function
print(arr1)
# mean
mean_arr1 = np.mean(arr1)
print(mean_arr1)
 
# Getting sum 
sum_arr1 = np.sum(arr1)
print(sum_arr1)
 
# Getting variance
var_arr1 = np.var(arr1)
print(var_arr1)
```

#### summation of matrix row and column wise

```python
arr1 = np.arange(1,13).reshape(3,4)
print(arr1)
```

```python

print ("The column wise summation of all matrix  is : ") 
print (np.sum(arr1,axis=0)) 
  
# using sum(axis=1) to print summation of all columns of matrix 
print ("The row wise summation of all matrix  is : ") 
print (np.sum(arr1,axis=1)) 
```

#### cumulative sum of array

```python
arr3 = [2,6,7,12]
np.sum(arr3)
```

```python
np.add.accumulate(arr3) # cumulative sum
```

```python
np.subtract.accumulate(arr3) # cumulative subtraction
```

```python
arr1 = np.arange(1,13).reshape(3,4)
print(arr1)
```

```python
np.add.accumulate(arr1) # cumulative sum, by default sum is along rows
```

```python
np.add.accumulate(arr1,axis = 1) # cumulative sum, axis = 1 sum is along columns
```

### stacking

#### hstack

hstack will join two or more arrays horizontally

```python
arr1 = np.array([2,8,10,12])
arr2 = np.array([10,20,30,40])

np.hstack((arr1,arr2)) 
```

```python
arr1 = np.arange(1,5).reshape((2,2))
arr2 = np.arange(10,50,10).reshape((2,2))
```

```python
print(arr1)
```

```python
print(arr2)
```

```python
np.hstack((arr1,arr2)) # hstack will increase the columns
```

```python
# we can join 2 or more arrays in horizontal way
arr1 = np.arange(1,5).reshape((2,2)) 
arr2 = np.arange(10,50,10).reshape((2,2))
arr3 = np.arange(100,700,100).reshape((2,3))
```

```python
np.hstack((arr1,arr2,arr3))
```

#### Vstack

```python
# vertical stacking the arrays
```

```python
arr1 = np.array([2,8,10,12])
arr2 = np.array([10,20,30,40])

np.vstack((arr1,arr2))  # vertical stack the arrays, hence increasing the rows
```

```python
arr1 = np.arange(1,5).reshape((2,2))
arr2 = np.arange(10,50,10).reshape((2,2))

np.vstack((arr1,arr2))
```

```python
# we can join 2 or more arrays in vertical way
arr1 = np.arange(1,5).reshape((2,2)) 
arr2 = np.arange(10,50,10).reshape((2,2))
arr3 = np.arange(100,500,100).reshape((2,2))
```

```python
np.vstack((arr1,arr2,arr3))
```
